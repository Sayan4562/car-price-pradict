## Second Aproch   Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
on=OneHotEncoder()
on.fit_transform(x[['car_name','registration_year','insurance_validity',
          'fuel_type','seats','kms_driven','ownsership',
          'transmission','mileage(kmpl)','engine(cc)','torque(Nm)']])
```

```python
colum_trnas=make_column_transformer((OneHotEncoder(categories=on.categories_),['car_name','registration_year',
                                                                               'insurance_validity','fuel_type','seats',
                                                                               'kms_driven','ownsership','transmission',
                                                                               'mileage(kmpl)','engine(cc)','torque(Nm)']),
                                                                                remainder='passthrough')
colum_trnas
```

```python
scores=[]
for i in range(300):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(colum_trnas,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
    
    
```

```python
max(scores)
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(colum_trnas,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
scores.append(r2_score(y_test,y_pred))
    
```

```python
import pickle

pickle.dump(pipe, open('LinerRegration_model.pkl','wb'))
```
